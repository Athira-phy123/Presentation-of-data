```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data1=pd.read_csv('cars_data.csv')
data1
```

```python
data1.isnull().sum()
```

insight : There is no null values in the data set

```python
data2=data1["Buyer Gender"].value_counts().to_frame().reset_index()
data2
```

```python
plt.bar(data2["index"],data2["Buyer Gender"],color=['blue','green'],width=0.3)
plt.title("Bar Graph of Male vs Female")
plt.xlabel("Buyers")
plt.ylabel("Count")
```

insight:The Number of Female buyers is more than Male buyers

```python
data1.groupby(by="Make")["Sale Price"].max().head()
```

Insight:The car Acura have highest Sale Price
